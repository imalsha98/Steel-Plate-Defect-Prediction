```python
from sklearn.model_selection import train_test_split
```

```python
X = train.iloc[:,:-7]
y = train.iloc[:,-7:]
X_test = test
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1,random_state=42)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from sklearn.multioutput import MultiOutputClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
n_estimators = 50
model = MultiOutputClassifier(RandomForestClassifier(n_estimators=n_estimators, random_state=42))
model.fit(X_train, y_train)
print("Training accuracy:", model.score(X_train, y_train))
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# Generate and visualize confusion matrices for each target class
classes = y.columns

for i, class_name in enumerate(classes):
    cm = confusion_matrix(y_test.iloc[:, i], y_pred[:, i])
    plt.figure(figsize=(10, 7))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix for {class_name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

    # Print classification report
    print(f'Classification Report for {class_name}:\n')
    print(classification_report(y_test.iloc[:, i], y_pred[:, i]))
```

```python
# Calculate the correlation matrix
corr_matrix = train.corr()

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Training Features')
plt.show()
```

```python
proba = model.predict_proba(X_test)
proba
```

```python
# Extract probabilities for each class and pad 
proba_pastry_padded = np.pad(proba[0][:, 1], (0, len(submission) - len(proba[0][:, 1])), 'constant')
proba_z_padded = np.pad(proba[1][:, 1], (0, len(submission) - len(proba[1][:, 1])), 'constant')
proba_k_padded = np.pad(proba[2][:, 1], (0, len(submission) - len(proba[2][:, 1])), 'constant')
proba_stain_padded = np.pad(proba[3][:, 1], (0, len(submission) - len(proba[3][:, 1])), 'constant')
proba_dirt_padded = np.pad(proba[4][:, 1], (0, len(submission) - len(proba[4][:, 1])), 'constant')
proba_bump_padded = np.pad(proba[5][:, 1], (0, len(submission) - len(proba[5][:, 1])), 'constant')
proba_other_padded = np.pad(proba[6][:, 1], (0, len(submission) - len(proba[6][:, 1])), 'constant')

# Assign the padded probabilities to the submission DataFrame
submission['Pastry'] = proba_pastry_padded
submission['Z_Scratch'] = proba_z_padded
submission['K_Scratch'] = proba_k_padded
submission['Stains'] = proba_stain_padded
submission['Dirtiness'] = proba_dirt_padded
submission['Bumps'] = proba_bump_padded
submission['Other_Faults'] = proba_other_padded

# Optionally, remove the last column if it's redundant
submission = submission.iloc[:, :-1]

submission


```
